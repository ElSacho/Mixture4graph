```python
import json
import networkx as nx

# Chemin vers votre fichier JSON
file_path = 'data/evelinag-star-wars-network/starwars-episode-7-mentions.json'

# Lire les données depuis le fichier JSON
with open(file_path, 'r') as file:
    data = json.load(file)

# Créer un graphique vide
star_wars = nx.Graph()

# Ajouter des nœuds au graphique
node_names = []  # Garder une trace de l'ordre des noms de nœuds
for node in data['nodes']:
    star_wars.add_node(node['name'], value=node['value'], colour=node['colour'])
    node_names.append(node['name'])

# Ajouter des arêtes au graphique
for edge in data['links']:
    source_name = node_names[edge['source']]
    target_name = node_names[edge['target']]
    star_wars.add_edge(source_name, target_name, weight=edge['value'])
```

```python
star_wars_data = data_visualisation(star_wars)
star_wars_data.plot_figure_graph()
star_wars_data.print_generalitize()
```

```python
model = mixtureModel(star_wars.subgraph(max(nx.connected_components(star_wars), key=len)))
model.fit(range(2,6))
model.plot_icl()
model.plot_jrx()
# model.plot_all_adjency_matrices(show_names = True)
```

```python
model = mixtureModel(star_wars.subgraph(max(nx.connected_components(star_wars), key=len)), initilisation_method='random')
model.fit(range(3,6))
model.plot_icl()
model.plot_jrx()
# model.plot_all_adjency_matrices(show_names = True)
```
